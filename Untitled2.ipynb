```python
import torch
from peft import PeftConfig, PeftModel
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig
```

```python
peft_model_id="mrm8488/falcon-7b-ft-codeAlpaca_20k-v2"
config=PeftConfig.from_pretrained(peft_model_id)
#model=AutoModelForCausalLM.from_pretrained(config.base_model_name_or_path, return_dict=True, load_in_8bit=True, device_map={"":0},trust_remote_code=True)
tokenizer=AutoTokenizer.from_pretrained(peft_model_id)
tokenizer.pad_token = tokenizer.eos_token
```

```python
DEVICE = "cuda:0"

prompt = f"""
: How to improve sales?
:
""".strip()

encoding = tokenizer(prompt, return_tensors="pt").to(DEVICE)
```

4 bit

```python
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16,
)
```

```python
# MODEL_ID = "curiousily/falcon-7b-qlora-chat-support-bot-faq"
#model=AutoModelForCausalLM.from_pretrained(config.base_model_name_or_path, return_dict=True, load_in_8bit=True, device_map={"":0},trust_remote_code=True)
config = PeftConfig.from_pretrained(peft_model_id)

model = AutoModelForCausalLM.from_pretrained(
    config.base_model_name_or_path,
    return_dict=True,
    device_map="auto",
    trust_remote_code=True,
    load_in_4bit=True,
)

model = PeftModel.from_pretrained(model, peft_model_id)
```

```python
generation_config = model.generation_config
generation_config.max_new_tokens = 200
generation_config.temperature = 0
generation_config.num_return_sequences = 1
generation_config.pad_token_id = tokenizer.eos_token_id
generation_config.eos_token_id = tokenizer.eos_token_id
```

```python
# %%timeit -r 5

# with torch.inference_mode():
#     outputs = model.generate(
#         input_ids=encoding.input_ids,
#         attention_mask=encoding.attention_mask,
#         generation_config=generation_config,
#         do_sample=False,
#         use_cache=True,
#     )
```

```python
with torch.inference_mode():
    outputs = model.generate(
        input_ids=encoding.input_ids,
        attention_mask=encoding.attention_mask,
        generation_config=generation_config,
        do_sample=False,
        use_cache=True,
    )

print(tokenizer.decode(outputs[0], skip_special_tokens=True))
```

```python
# MODEL_ID = "curiousily/falcon-7b-qlora-chat-support-bot-faq"
config = PeftConfig.from_pretrained(peft_model_id)

model = AutoModelForCausalLM.from_pretrained(
    config.base_model_name_or_path,
    return_dict=True,
    device_map="auto",
    trust_remote_code=True,
    load_in_8bit=True,
)

model = PeftModel.from_pretrained(model, peft_model_id)
```

```python
generation_config = model.generation_config
generation_config.max_new_tokens = 200
generation_config.temperature = 0
generation_config.num_return_sequences = 1
generation_config.pad_token_id = tokenizer.eos_token_id
generation_config.eos_token_id = tokenizer.eos_token_id
```

```python
with torch.inference_mode():
    outputs = model.generate(
        input_ids=encoding.input_ids,
        attention_mask=encoding.attention_mask,
        generation_config=generation_config,
        do_sample=False,
        use_cache=True,
    )

print(tokenizer.decode(outputs[0], skip_special_tokens=True))
```
